```python
#Removes columns that don't exist in both datasets.
def removeVoidColumns(train,test):
  for i in train.columns:
    if i not in test.columns:
      train = train.drop(i,axis = 1)
  return train
```

```python
#Combines two DataFrames into one big frame
def combineFrames(frames):
  output = pd.DataFrame({})
  for i in frames:
    for j in i.columns:
      output[j] = i[j]
  return output
```

```python
#Gets the requisite integer features
def getfeatures(train_data,train_features_count):
  corr_matrix = train_data.corr()
  correlated_data = corr_matrix['SalePrice'].sort_values(ascending = False)
  return correlated_data.index[1:train_features_count+1].tolist()
```

```python
#Returns the mean of the weight if it exists, if it doesn't, returns 1
def findmean(index,data_means):
  if index >= len(data_means):
    return 1
  return data_means[index]
```

```python
#Returns the means of each column in a DataFrame
def getmean(data):
  means = []
  for i in range(len(data.columns)):
    total = 0
    count = 0
    for j in data.iloc[:,i]:
      total += j
      count += 1
    means.append(total/count)
  return means
```

```python
#Main function for canoicalization, returns each datapoint averaged by the mean.
def canonicalization(data,means):
  output = pd.DataFrame({})
  for i in range(len(data.columns)):
    mean = means[i]
    averaged_data = data.iloc[:,i] / mean
    output[data.columns[i]] = averaged_data
  return output
    
```

```python
#Finds the number and type of unique strings in a column
def unique_strings(column):
  unique_strings = []
  for i in column:
    if i not in unique_strings:
      unique_strings.append(i)
  return unique_strings
```

```python
#Changes string columns into an integer DataFrame based on the number of unique strings
def changeStrings(column):
  output = {}
  strings = unique_strings(column)
  for i in strings:
    output[str(column.name)+"_"+str(i)] = []
  for i in column:
    for j in strings:
      if i == j:
        output[str(column.name)+"_"+str(j)].append(1)
      else:
        output[str(column.name)+"_"+str(j)].append(0)
  return pd.DataFrame(output)
```

```python
#Main function to change a DataFrame of strings into an integer dataframe.
def changeFeatures(data):
  frames = []
  output = pd.DataFrame({})
  for i in range(len(data.columns)):
    columnFrame = changeStrings(data.iloc[:,i])
    frames.append(columnFrame)
  
  output = combineFrames(frames)
  return output
```

```python
#Calculates cost function
def mean_squared_error(y_true, y_predicted):
    ydelta = y_true-y_predicted
    cost = np.sum((y_true-y_predicted)**2) / len(y_true)
    return cost
```

```python
#Individual weights for each variable
def appendWeights(length):
  weights = []
  for i in range(length):
    weights.append(0.1)
  return weights
```

```python
#Predicts output based on weights and bias for x.
def predictY(current_bias, current_weights, x):
  y_predicted = current_bias
  for i in range(len(x.columns)):
    y_predicted = y_predicted + current_weights[i] * x.iloc[:,i]
  return y_predicted
```

```python
#Updates the bias (Theta 0)
def updateCurrentBias(current_bias,learning_rate,y_predicted,y,n):
  return current_bias - learning_rate * sum(y_predicted-y) / n
```

```python
#Updates the weights (Theta i)
def updateCurrentWeights(current_weights,learning_rate,y_predicted,y,n,x):
  for i in range(len(x.columns)):
    current_weights[i] = current_weights[i] - (learning_rate * sum((y_predicted - y) * x.iloc[:,i]) / n)
  return current_weights
```

```python
#Main gradient descent function
def multivariate_gradient_descent(x, y, iterations = 1000, learning_rate = 0.07,
                     stopping_threshold = 1e-6):
  y_comparison = pd.DataFrame(y)
  #Weights and Bias
  current_weights = appendWeights(len(x.columns))
  current_bias = 0.01

  #Length of dataset
  n = float(len(y))
  
  #Tracking costs & weights for debugging
  costs = []
  weights = []
  previous_cost = None

  #Iterating through the given iterations
  for i in range(iterations):

    #Predicting Y based on given weights and bias (Theta 0, Theta i)
    y_predicted = predictY(current_bias,current_weights,x)
    current_cost = mean_squared_error(y,y_predicted)

    #If little or no change, stop
    if previous_cost and abs(previous_cost - current_cost) <= stopping_threshold:
      break
    #Tracking weights/cost
    previous_cost = current_cost
    costs.append(current_cost)
    weights.append(current_weights)

    #Updating bias and weights
    current_bias = updateCurrentBias(current_bias, learning_rate, y_predicted, y, n)
    current_weights = updateCurrentWeights(current_weights,learning_rate,y_predicted,y,n,x)

    #Printing statement of iterations, costs, current weights and current bias
    print(f"Iteration {i+1}: Cost {current_cost}, Weight \{current_weights}, Bias {current_bias}")
    y_comparison['y-predicted'] = y_predicted
    #print(y_comparison)
  plt.plot(costs)
  plt.show()
  #Returning optimized weights and bias
  return current_weights,current_bias
```
